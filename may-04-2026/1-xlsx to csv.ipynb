```python
import pandas as pd

def convert_xlsx_to_csv(input_file, output_file):
    try:
        # Load the xlsx file
        # 'sheet_name=0' loads the first sheet
        df = pd.read_excel(input_file, sheet_name=0)

        # Export to CSV
        # 'index=False' prevents pandas from writing row numbers to the CSV
        df.to_csv(output_file, index=False, encoding='utf-8')
        
        print(f"Success! '{input_file}' has been converted to '{output_file}'.")
    
    except Exception as e:
        print(f"An error occurred: {e}")

# Usage
convert_xlsx_to_csv('province.xlsx', 'province.csv') #this is where the name of the input and output filw name
```
